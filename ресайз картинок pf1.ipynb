```python
for i in range(len(df)):
    response = requests.get(df.loc[i,'URL']) # посылаем запрос к урл из списка урл
    if response.status_code == 200:
            image = Image.open(BytesIO(response.content)) # получаем контент запроса
            df.loc[i, 'picture_name']= df.loc[:,'URL'].apply(lambda x: x.split( '/'))[i][-1]
            df.loc[i, 'width'] = image.size[0]  # Определяем ширину контента запроса
            df.loc[i, 'height'] = image.size[1]  # Определяем высоту контента запроса
            df.loc[i, 'gross'] = df.loc[i, 'Content-Length']
```

```python
df
```

```python
%%time
for i in range(len(df)):
    width = df.loc[i, 'width']
    height = df.loc[i, 'height']
    gross = df.loc[i, 'gross']
    while gross > 100000 and  width > 500 and height > 500:
            width = int(width/2)
            height = int(height/2)
            try:
                resized_image = image.resize((width, height), Image.BICUBIC) 
                resized_image.save(df.loc[i,'picture_name']) # сохраняем на локальном компьютере
            except:
                print('error',i)
            df.loc[i, 'new_gross']=gross
            df.loc[i, 'new_width'] = width
            df.loc[i, 'height'] = height
```
